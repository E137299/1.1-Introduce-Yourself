#**ALGORITHMS and DECISION MAKING**

An **algorithm** is a step-by-step procedure, which defines a set of instructions to be executed in a certain order to get the desired output.

**Characteristics of an Algorithm**
Not all procedures can be called an algorithm. An algorithm should have the following characteristics −

- **Unambiguous** − Algorithm should be clear and unambiguous. Each of its steps (or phases), and their inputs/outputs should be clear and must lead to only one meaning.
- **Input** − An algorithm should have 1 or more well-defined inputs.
- **Output** − An algorithm should have 1 or more well-defined outputs, and should match the desired output.
- **Finiteness** − Algorithms must terminate after a finite number of steps.
- **Feasibility** − Should be feasible with the available resources.


**if** statement is the most simple decision-making statement. It is used to decide whether a certain statement or block of statements will be executed or not i.e if a certain condition is true then a block of statement is executed otherwise not.

When the if statement executes, the condition is tested, and if it is true the block statements are executed. otherwise, block statements are skipped entirely.

Python relies on indentation (whitespace at the beginning of a line) to define the block statements. In the example below, statement1, statement2 and statement3 are indented and will be executed if the condition is true. statement4 will be executed regardless of whether the condition is True or False because it is not part of the block statement.

```
if condition:
   statement1
   statement2
   statement3
statement4
```
if.PNG


````python
age = int(input("How old are you?  "))

if(age >= 18): #condition
  # if block. These statements will be executed if the condition evaluates to True.
  print("\nYou are eligible to vote!")
  print("You can buy cigarettes!")
  print("You can be called for jury duty!")

#The following statement will be executed regardless of the condition.
print("Goodbye")
````

#**if/else STATEMENTS**

The if statement alone tells us that if a condition is true it will execute a block of statements and if the condition is false it won’t. But what if we want to do something else if the condition is false. Here comes the if/else statement. We can use the else statement with if statement to execute a block of code when the condition is false.

if/else statements are dual alternative decision structure. The program executes one of two possible paths. The if block is taken if the condition is True, and the else block is executed if the condition evaluates to False.

**Note:** Remember that indentation defines the blocks(statements to be executed).

**Note:** The if and else must be aligned vertically.



```
if (condition):
    # if block
    statement1
    statement2
    statement3
else:
    # else block
    statement4
    statement5
    statement6
statement7
```

if else.PNG


````python
choice = input("Would like a toast or cereal for breakfast?  ")

if choice == "toast": #condition
  # if block. This will be executed if the condition evaluates to True
  print("\nToast!")
  print("Step 1: Place a slice of bread in the toaster.")
  print("Step 2: Remove the toast from the toaster.")
  print("Step 3: Put butter and jelly on the toast.")

else:
  # else block. This will be executed if the condition evaluates to False
  print("\nCereal!")
  print("Step 1: Get a bowl and milk, and select your cereal.")
  print("Step 2: Pour the cereal into the bowl.")
  print("Step 3: Pour milk into the bowl with the cereal.")

#The following statement will be executed regardless of the condition.
print("Step 4: Sit at the table and enjoy your breakfast.")
````

#**if/elif/else STATEMENTS**

The **elif** statement allows you to check multiple expressions for True and execute a block of code as soon as one of the conditions evaluates to True.

Similar to the else, the elif statement is optional. However, unlike else, for which there can be at most one statement, there can be an arbitrary number of elif statements following an if.

**Important:**  As soon as one of the conditions controlling the if is true, the statement associated with that if is executed, and the rest of the ladder is bypassed. If none of the conditions is true, then the final else statement will be executed.



```
if (condition1):
    statement1
    statement2
    statement3
elif (condition2):
    statement4
    statement5
    statement6
elif (condition3)
    statement7
    statement8
    statement9
.
.
else:
    statement10
```

````python
choice = input("Would like toast, cereal, scrambled eggs or a bagel for breakfast?  ")

if choice == "toast": # first condition
  # if block. This will be executed if the condition evaluates to True
  print("\nToast!")
  print("Step 1: Place a slice of bread in the toaster.")
  print("Step 2: Remove the toast from the toaster.")
  print("Step 3: Put butter and jelly on the toast.")

elif choice == "cereal": # second condition
  # first elif block. This will be executed if the second condition evaluates to True
  print("\nCereal!")
  print("Step 1: Get a bowl and milk, and select your cereal.")
  print("Step 2: Pour the cereal into the bowl.")
  print("Step 3: Pour milk into the bowl with the cereal.")

elif choice == "scrambled eggs": # third condition
  # second elif block. This will be executed if the third condition evaluates to True
  print("\nEggs!")
  print("Step 1: Break a couple of eggs into bowl.")
  print("Step 2: Vigorously wisk the eggs with a bit of milk and pepper.")
  print("Step 3: Cook the eggs in a skillet. Remember to continously scramble the eggs.")

elif choice == "bagel": # fourth condition
  # third elif block. This will be executed if the fourth condition evaluates to True
  print("\nBagel!")
  print("Step 1: Halve the bagel down its length with a knife. Be careful.")
  print("Step 2: Toast the bagel in the toaster.")
  print("Step 3: Spread cream cheese on each half of the bagel.")

else:
  # else block. This will be executed if all of the conditions above evaluate to False. This is the default block when all of the conditions are False.
  print("\nError!")
  print("Your choice did not match any of the options.")
  print("Take a granola bar from the pantry.")
  print("Skip to Step 4.")


#The following statement will be executed regardless of the condition.
print("Step 4: Sit at the table and enjoy your breakfast.")
````

**Logical operators** are used to combine conditional statements:

|OPERATOR|DESCRIPTION|EXAMPLE|
|---|---|---|
|and|Returns True if both statements on either side of the 'and' are True| x >= 0 and x <= 10 |
|or| Returns True if at least one of the statements on either side of the 'or' is True.| fav_dessert == "ice cream" or fav_dessert == "cookies"|
|not|Reverse the result, returns False if the result is True| not(x >= 0 and x <= 10)|

**Logical Operator: and**

````python
clean = input("Have you cleaned your room? (y/n) ")
homework = input("Have you completed your homework? (y/n)")

if clean=="y" and homework =="y": # compound condition. If both conditions on either side of the 'and' are True, the compound condition will evaluate to True.
  print("\nYou can go to the party, because you have cleaned your room and completed your homework.")
else:
  print("\nYou cannot go to the party. You need to either clean your room or complete your homework in order or both.")
````

**Logical Operator: or**

````python
can_drive = input("Can you drive yourself to the party? (y/n)")
friend_with_car = input("Do you have a friend who can drive you to the party? (y/n)")

if can_drive == "y" or friend_with_car == "y":
  print("\nYou can go to the party, because you can either drive yourself or you have a friend who can drive you.")
else:
  print("\nUnfortunately, you cannot go to the party, because you cannot get a ride to the party.")
````

**Logical Operator: not**

````python
boogers = input("Do you pick your nose? (y/n)")

if not(boogers=='y'):
  print("\nCool")
else:
  print("\nEew")
````

###**NESTED if STATEMENTS**

There may be a situation when you want to check for another condition after a condition resolves to true. In such a situation, you can use the nested if construct.

In a nested if construct, you can have an if...elif...else construct inside another if...elif...else construct.



```
if condition1:
    # if block1
    if condition2: # this condition will only be evaluated by the program if condtion1 is True
        # if block2 will be executed if condition1 and condition2 are both True
        statement1
        statement2
    else:
        # else block2 will be excuted if condition1 is True and condition2 is False
        statement3
        statement4
else:
    # else block1 will be executed if condition1 is False
    statement5
    statement6
```



Any number of these statements can be nested inside one another. Indentation is the only way to figure out the level of nesting. They can get confusing, so they must be avoided unless necessary.
